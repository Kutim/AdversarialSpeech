```python
from InverseMfcc import InverseMfcc
import librosa
import numpy as np
```

# Convert audio MFCC and back

```python
y, sr = librosa.load('audios/p271/p271_089.wav', mono=True, sr=None)
y = y[::3]
imfcc = InverseMfcc(y, 16000)
imfcc.transform()
```

```python
imfcc.play_original()
```

```python
imfcc.play_transformed()
```

```python
imfcc.modify_parameters(noise_scale=.1)
imfcc.transform()
imfcc.play_transformed()
```

```python
imfcc.modify_parameters(noise_scale=0.0, mfcc_noise=10.0)
imfcc.transform()
imfcc.play_transformed()
```

# Update Transform parameters (same input)

```python
# Update the transform parameters
imfcc.modify_parameters(num_mfcc=150, include_deltas=True)
imfcc.transform()
imfcc.play_transformed()
```

# Set MFCC directory (original)

```python
# Transform original numpy file to same shape
orig = np.load("orig_x_no_aug.npy", allow_pickle=False)
curr = orig.transpose([2, 0, 1])[:,0,:]

# Used to remove trailing zeros to align dimensions
for i in range(curr.shape[1]):
    if ((curr[:,i] == 0).all()):
        break
curr = curr[:,:i]

imfcc.transform(curr)
imfcc.play_transformed()
```

# Set MFCC from Adversarial Version

```python
# Transform original numpy file to same shape
adv = np.load("adv_x.npy", allow_pickle=False)
curr = adv.transpose([2, 0, 1])[:,0,:]

# Used to remove trailing zeros to align dimensions
curr = curr[:,:71] ## will need to track where zero padding starts!

imfcc.transform(curr)
imfcc.play_transformed()
```
